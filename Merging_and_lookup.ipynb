**Data Preparation and Missing Value Handling**

Here 3 dataframe is created.
check is null value is occured in each column.
compute percentage of null value .
Then fill city name having null with "Unknown" .
And drop the row which name is null

```python
import pandas as pd
students_data = {
    'student_id': [101, 102, 103, 104, 105, 106, 107],
    'name': ['Alice', 'Bob', None, 'David', 'Emma', 'Frank', 'Grace'],
    'email': ['alice@email.com', 'bob@email.com', 'charlie@email.com', None, 'emma@email.com', 'frank@email.com', 'grace@email.com'],
    'city': ['Mumbai', 'Delhi', 'Bangalore', 'Mumbai', None, 'Chennai', 'Delhi']
}
df_stud=pd.DataFrame(students_data)
print(df_stud)


enrollments_data = {
    'student_id': [101, 102, 103, 105, 108, 109],
    'course_name': ['Python', 'Data Science', 'Python', 'Machine Learning', 'AI', 'Python'],
    'enrollment_date': ['2024-01-15', '2024-01-20', '2024-02-01', '2024-02-10', '2024-02-15', '2024-03-01']
}
df_enrol=pd.DataFrame(enrollments_data)
print(df_enrol)


scores_data = {
    'student_id': [101, 102, 104, 105, 106],
    'exam_score': [85, 92, 78, 88, 95]
}
df_score=pd.DataFrame(scores_data)
print(df_score)
```

```python
print(df_stud.isnull().sum())
print(df_stud.isnull().sum()/len(df_stud)*100)
```

```python
df_stud["city"]=df_stud["city"].fillna("Unknown")
df_stud=df_stud.dropna(subset=["name"])
print(df_stud.reset_index(drop=True))
```

**Multiple Join Operations**

Multiple join is performed
isin()-- is used to check it is in the desired table
isna()--is used to identify if it has any nan value


```python
pd.set_option('display.max_columns', 1000)
pd.set_option('display.width', 1000)
inner=pd.merge(df_stud,df_enrol,on="student_id",how="inner")
print("Inner Join Result ",len(inner),"rows")
print(inner)
excluded_students=df_stud[~df_stud["student_id"].isin(df_enrol["student_id"])]["student_id"].tolist()
print("Excluded students:",excluded_students)
```

```python
left=pd.merge(df_stud,df_enrol,on="student_id",how="left")
print("Left Join Result",len(left),"rows")
print(left)
null_course=left[left["course_name"].isna()]["student_id"].tolist()
print("Students with null course_name:",null_course)
```

```python
right=pd.merge(df_stud,df_enrol,on="student_id",how="right")
print("Left Join Result",len(right),"rows")
print(right)
no_name=right[right["name"].isna()]["student_id"].tolist()
print("Student with no name:",no_name)
```

```python
outer=pd.merge(df_stud,df_enrol,on="student_id",how="outer",indicator=True)
print("Left Join Result",len(outer),"rows")
print(outer)
print(outer["_merge"].value_counts())
either=outer[(outer["name"].isna())|(outer["course_name"].isna())]
print("rows with null of either name or course\n",either)
```

```python
score_dict=dict(zip(df_score["student_id"],df_score["exam_score"]))
df_stud["exam_score"]=df_stud["student_id"].map(score_dict)
print(df_stud)
```

```python
merge_left=pd.merge(df_stud,df_score,on="student_id",how="left")
print(merge_left)
```

**Lookup Operation and Automation**

Create a function to perform a different joins.
It pass the join type,and name of column has a common in both table.


```python
def auto_merge(df1, df2, join_type, key_column):
  merge_operation=pd.merge(df1, df2,on=key_column,how=join_type)
  result={
      'result_df': merge_operation,
      'row_count': len(merge_operation),
      'join_type': join_type
  }
  return result
test1=auto_merge(df_stud,df_enrol,"outer","student_id")
print("Automation Function Test:")
print("Join Type:",test1["join_type"])
print("Rows in Result:",len(test1))
print(test1)
test2=auto_merge(df_stud,df_enrol,"left","student_id")
print("Automation Function Test:")
print("Join Type:",test2["join_type"])
print("Rows in Result:",len(test2))
print(test2)
```
